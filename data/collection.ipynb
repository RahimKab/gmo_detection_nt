```python
from Bio import Entrez, SeqIO
Entrez.email = "your_email@example.com"
```

```python
handle = Entrez.esearch(
    db="nucleotide",
    term="Gossypium hirsutum[Organism]"
)
record = Entrez.read(handle)
handle.close()

ids = record["IdList"]
```

```python
print(ids)
```

```python
sequences = []

for seq_id in ids:
    handle = Entrez.efetch(
        db="nucleotide",
        id=seq_id,
        rettype="fasta",
        retmode="text"
    )
    seq_record = SeqIO.read(handle, "fasta")
    handle.close()
    sequences.append(seq_record)
    
print(f"Fetched {len(sequences)} sequences.")
SeqIO.write(sequences, "Gossypium hirsutum.fasta", "fasta")
```

```python
handle = Entrez.esearch(
    db="nucleotide",
    term="cry1Ac OR cry2Ab",
    retmax=20
)
record = Entrez.read(handle)
handle.close()

bt_ids = record["IdList"]
```

```python
bt_sequences = []

for seq_id in bt_ids:
    handle = Entrez.efetch(
        db="nucleotide",
        id=seq_id,
        rettype="fasta",
        retmode="text"
    )
    seq_record = SeqIO.read(handle, "fasta")
    handle.close()
    bt_sequences.append(seq_record)

SeqIO.write(bt_sequences, "raw/bt_genes.fasta", "fasta")
```
